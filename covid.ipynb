```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('synthetic_covid19_data.csv')
df
```

```python
#cleaning data
df.isnull().sum()
```

```python
n = ['population_density','median_age',
    'aged_65_older','aged_70_older',
    'gdp_per_capita','extreme_poverty',
    'cardiovasc_death_rate','diabetes_prevalence'] #droping columns

df = df.drop(n, axis=1)
df
```

```python
# 1. Which continent reported the highest total deaths?

continent_summary = df.groupby('continent')[['total_cases', 'total_deaths']].max().reset_index()
continent_summary

sns.lineplot(x='continent',y='total_deaths',data=continent_summary)
```

```python
# 2. Which continent reported the highest total cases?

continent_summary = df.groupby('continent')[['total_cases', 'total_deaths']].max().reset_index()
continent_summary

sns.lineplot(x='continent',y='total_cases',data=continent_summary)
```

```python

continent_summary = df.groupby('continent')[['total_cases', 'total_deaths']].max().reset_index()
continent_summary

sns.lineplot(x='total_cases',y='total_deaths',data=continent_summary)

#Based on continent-wise maximum totals,
#  there isn’t a clear linear relationship between total cases and total deaths.
#  Some continents with fewer total cases report similar or higher deaths, 
# possibly due to differences in population size, healthcare, or reporting practices.
```

```python

plt.figure(figsize=(12,5))

# Total cases by continent
plt.subplot(1,2,1)
sns.barplot(x='continent', y='total_cases', data=df,color='green')
plt.title('Total COVID-19 Cases by Continent')

# Total deaths by continent
plt.subplot(1,2,2)
sns.barplot(x='continent', y='total_deaths', data=df,color='red') 
plt.title('Total COVID-19 Deaths by Continent')

plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(18,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of COVID-19 Features')
plt.show()
```

```python
# 4. Which continent conducted the most total tests?	
ds = df.groupby('continent')[['total_tests','total_deaths']].max()
print(ds)
sns.lineplot(x='continent',y='total_tests',data=ds,color='orange')
```

```python
#  5. How does vaccination progress differ by continent?
v = df.groupby('continent')['people_fully_vaccinated'].max().reset_index()
print(v)

plt.figure(figsize=(10,6))
sns.lineplot(x='continent', y='people_fully_vaccinated', data=v, marker='o', color='green')
plt.title('Fully Vaccinated People by Continent')
plt.xlabel('Continent')
plt.ylabel('People Fully Vaccinated')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# 6. Which continent had the highest test positivity rate?

po = df.groupby('continent')['positive_rate'].mean().reset_index()
print(po)

sns.scatterplot(x='continent',y='positive_rate',data=po,color='red')
```

```python
# 7.What’s the average stringency index per continent?

st = df.groupby('continent')['stringency_index'].mean().reset_index()
print(st)


sns.jointplot(x='continent', y='stringency_index', data=st)
```

```python
# 8. Which 10 countries recorded the highest total cases?


# Top 10 countries with highest total cases
co = df.groupby('location')['total_cases'].max().nlargest(10).reset_index()

colors = [
    'red', 'blue', 'green', 'orange', 'purple',
    'pink', 'yellow', 'cyan', 'magenta', 'brown'
]

# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(
    co['total_cases'],
    labels=co['location'],
    colors=colors,              
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}
)

plt.title('Top 10 Countries with Highest Total COVID-19 Cases', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 8. Which 10 countries recorded the highest total cases?


# Top 10 countries with highest total cases
co = df.groupby('location')['total_cases'].max().nlargest(10).reset_index()

sns.barplot(x='location',y='total_cases',data=co,color='red')
```

```python
# 9. Which 10 countries recorded the most deaths?

de = df.groupby('location')['total_deaths'].max().nlargest(10).reset_index()

print(de)


sns.scatterplot(x='location',y='total_deaths',data=de,color='purple')
```

```python
positive = df.groupby('population')['positive_rate'].max().reset_index()
print(positive)

plt.figure(figsize=(11,6))
sns.histplot(x='population',y='positive_rate', data=df, edgecolor='black',color='red')
plt.xticks()
plt.tight_layout()
plt.show()
```
